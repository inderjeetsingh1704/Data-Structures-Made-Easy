```python
# bubble sort
def bubble_sort(arr):

    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
            
    return arr
```

```python
import random
```

```python
lists = [random.randint(0,10000) for _ in range(10000)]
```

```python
bubble_sort(lists)
```

```python
def insertion_sort(arr):

    for i in range(1,len(arr)):
        current_val = arr[i]
        j = i-1
        while j>=0 and arr[j]>current_val:
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = current_val
    return arr
        
```

```python
lists = [random.randint(0,1000000) for _ in range(1000000)]
insertion_sort(lists)
```

```python
def merge(left,right):
    if len(left)==0:
        return right
    if len(right) == 0:
        return left
    
    merged = []
    left_index = right_index = 0
    while len(merged) < len(left) + len(right):
        if left[left_index]<=right[right_index]:
            merged.append(left[left_index])
            left_index+=1
        else:
            merged.append(right[right_index])
            right_index+=1
    
        if right_index == len(right):
            merged.extend(left[left_index:])

        if left_index == len(left):
            merged.extend(right[right_index:])

    return merged
```

```python
def merge_sort(arr):

    if len(arr)<2:
        return arr

    midpoint = len(arr)//2
    return merge(merge_sort(arr[:midpoint]),merge_sort(arr[midpoint:]))
```

```python
lists = [random.randint(0,100000000) for _ in range(100000000)]
merge_sort(lists)
```

```python
def quicksort(arr):
    if len(arr)<2:
        return arr
    
    low,same,high = [], [],[]
    pivot = arr[random.randint(0,len(arr)-1)]

    for item in arr:
        if item <pivot:
            low.append(item)
        elif item ==arr:
            same.append(item)
        elif item >pivot:
            high.append(item)
    
    return quicksort(low) +same + quicksort(high)
```
